## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

url = "https://raw.githubusercontent.com/manthansingh26/FLY_Manthan/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

print("Total pages in dataset:", len(df))
print("\nColumns available:", list(df.columns[:10]), "... (and more)")

print("\nTrend direction counts:")
print(df["trend_direction"].value_counts(normalize=True).round(3))

print("\nMedian avg_position by trend_direction:")
print(df.groupby("trend_direction")["avg_position"].median())

print("\nMedian ctr by trend_direction:")
print(df.groupby("trend_direction")["ctr"].median())
```
